```python
from tkinter import *
import requests
root=Tk()
root.title('Weather prediction')
root.geometry('700x600')

cityvar=StringVar()
latvar=StringVar()
logvar=StringVar()
weathervar=StringVar()

Label(root,text="Enter the city").grid(row=0,column=0)
Entry(root,textvariable=cityvar).grid(row=0,column=1)

def getdetails():
    api_address = 'http://api.openweathermap.org/data/2.5/weather?appid=3b38f7256941d04dc249f965f535ca29&q='
    city = cityvar.get()
    url = api_address + city
    json_data = requests.get(url).json()
    #print(json_data)
    latvar.set(json_data['coord']['lat'])
    logvar.set(json_data['coord']['lon'])
    weathervar.set(json_data['weather'][0]['main'])



Button(root,text="submit",command=getdetails).grid(row=1,column=1)
Label(root,text="Latitude values").grid(row=2,column=0)
Entry(root,textvariable=latvar).grid(row=2,column=1)
Label(root,text="Longitude values").grid(row=3,column=0)
Entry(root,textvariable=logvar).grid(row=3,column=1)
Label(root,text="weather").grid(row=4,column=0)
Entry(root,textvariable=weathervar).grid(row=4,column=1)

root.mainloop()
```
